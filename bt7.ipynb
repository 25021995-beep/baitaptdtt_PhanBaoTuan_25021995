```python
def binsearch(arr, x):
    low = 0
    high = len(arr)-1
    while low<=high:
        mid=(low+high)//2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            low = mid+1
        elif arr[mid] > x:
            high = mid-1
    else:
        return -1


a=list(map(int,input().split()))
x=int(input())
print(binsearch(a,x))
```

```python
def pair(arr,a):
    count=0
    for i in range(len(arr)):
        for j in range(i+1,len(arr)):
            if arr[i]+arr[j]==a:
                count+=1
    return count

a=list(map(int,input().split()))
x=int(input())
print(pair(a,x))
```

```python
def flatten(arr):
    c=[]
    for i in range(len(arr)):
        for j in range(len(arr[i])):
            if i%2==0:
                c.append(arr[i][j])
            if i%2==1:
                arr[i].sort(reverse=True)
                c.append(arr[i][j])
    return c

a=input()
b=eval(a)
print(flatten(b))

```

```python
def longestInc(arr):
    c=[]
    for i in range(len(arr)):
        count=1
        a=arr[i]
        for j in range(i+1,len(arr)):
            if arr[i]<arr[j]:
                count+=1
                arr[i]=arr[j]
        arr[i]=a
        c.append(count)
    return max(c)
a=input()
b=eval(a)
print(longestInc(b))
```

```python
def interval(arr1,arr2):
    h=[]
    for i in arr2:
        d=[]
        for j in arr1:
            if j[0]<=i<=j[1]:
                c=j[1]-j[0]+1
                d.append(c)
        if d==[]:
            d.append(-1)
        h.append(min(d))
    return h

a=input()
b=eval(a)
c=input()
c=eval(c)
print(interval(b,c))
```

```python
def LnumRight(arr):
    b=[]
    for i in range(len(arr)):
        count=0
        for j in range(i+1,len(arr)):
            if arr[j]<arr[i]:
                count=count+1
        b.append(count)
    return b

a=input()
a=eval(a)
print(LnumRight(a))
```

```python
def mod(a,b):
    c=[]

    for i in range(len(a)):
        sum=0
        for j in range(i,len(a)):
            sum=sum+a[j]
            c.append(sum%b)
    return max(c)
a=input()
a=eval(a)
b=int(input())
print(mod(a,b))
```

```python
def find(a,key):
    count=0
    for i in range(len(a)):
        if len(key)==1 and a[i]==key[0] or a[i]==key[0]+"." or a[i]==key[0]+",":
            count=count+1
        elif len(key)>1 and a[i]==key[0] and a[i+1]==key[1]+".":
            count=count+1

    return count
a=map(str,input().split())
b=list(a)

key=input().split()
key=list(key)

print(b)
print(key)
print(find(b,key))
```

```python
def harmonicSeries(arr):
    d=[]
    for i in range(len(arr)):
        c1=[]
        c2=[]
        c1.append(arr[i])
        c2.append(arr[i])
        for j in range(i+1,len(arr)):
            if arr[i]>=arr[j] and (arr[i]==arr[j] or arr[i]==arr[j]+1):
                 c1.append(arr[j])
            elif arr[i]<=arr[j] and (arr[i]== arr[j] or arr[i]==arr[j]-1):
                c2.append(arr[j])
        if len(c1)>len(c2):
            d.append(len(c1))
        else:
            d.append(len(c2))

    return max(d)

a= list(map(int,input().split()))
print(harmonicSeries(a))
```

```python
def FindingNemo(arr):
    for i in range(len(arr)):
        if arr[i]=="Nemo" and i==len(arr)-1:
            return f"{arr[i-1]} and {arr[0]}"
        elif arr[i]=="Nemo" and i==0:
            return f"{arr[-1]} and {arr[i+1]}"
        elif arr[i]=="Nemo" and 0<i<len(arr)-1:
            return f"{arr[i-1]} and {arr[i+1]}"
n=int(input())
a=[]
while len(a)<n:
    c=input()
    a.append(c)
for i in range(len(a)):
    for j in range(len(a[i])):
        if a[i][j]==" ":
            a[i]=a[i][0:j]

print(a)
print(FindingNemo(a))
```

```python
def bubbleSort(a):
    n = len(a)
    for i in range(n-1):
        for j in range(i+1,n):
            if a[j]<a[i]:
                a[j],a[i]=a[i],a[j]
    return a
a=list(map(int,input().split()))
print(bubbleSort(a))
```
