```python
def MaxOfTwo(a,b):
    if a > b:
        return a
    else:
        return b
a,b=map(int,input().split())
print(MaxOfTwo(a,b))
```

```python
def swap(a,b):
    a,b=b,a
    c=(a,b)
    return c
a,b=map(int,input().split())
print(swap(a,b))
```

```python
def isPrime(n):
    i=2
    if n==2:
        return True
    elif n<2:
        return False
    while i<(n**0.5)+1:
        if n%i==0:
            return False
        i=i+1
    else:
        return True
a=int(input())
print(isPrime(a))
```

```python
def Perfect(a):
    i=1
    sum=0
    while i<(a**0.5)+1:
        if a%i==0:
            sum=sum+i
        i=i+1
    if sum==a:
        return True
    else:
        return False
a=int(input())
print(Perfect(a))
```

```python
def location(arr,k):
    i=0
    while i<len(arr):
        if arr[i]==k:
            return i+1
        i=i+1
    else:
        return -1

arr=list(map(int,input().split()))
k=int(input())
print(location(arr,k))
```

```python
def fructional(a):
    multi=1
    while a!=1:
        multi=multi*a
        a=a-1
    return multi
a=int(input())
print(fructional(a))
```

```python
def cal(a,b):
    sum=round(a+b,2)
    minus=round(a-b,2)
    divide=round(a/b,2)
    multi=round(a*b,2)
    return sum,minus,divide,multi
a,b=map(float,input().split())
```

```python
def hamming(a,b):
    c=a^b
    c=bin(c)[2:]
    i=0
    count=0
    while i<len(c):
        if c[i]=='1':
            count+=1
        i=i+1
    return count
a,b=map(int,input().split())
print(hamming(a,b))
```

```python
def sum(a):
    sum=0
    while a!=0:
        sum=a%10+sum
        a=a//10
    return sum
a=int(input())
print(sum(a))
```

```python
def word(a,b):
    i1=0
    arr1=[]
    while i1<len(a):
        j1=i1+1
        while j1<len(a):
            if a[i1]==a[j1]:
                arr1.append(i1)
                arr1.append(j1)
            j1=j1+1
        i1=i1+1
    i2=0
    arr2=[]
    while i2<len(b):
        j2=i2+1
        while j2<len(b):
            if b[i2]==b[j2]:
                arr2.append(i2)
                arr2.append(j2)
            j2=j2+1
        i2=i2+1
    if arr1==arr2:
        return True
    else:
        return False

a,b=input().split()
print(word(a,b))
```
