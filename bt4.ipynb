```python
#6
def isPrime(n):
    if n <= 1: return False
    elif n<=2: return True
    else:
        d = 2
        while n%d !=0:
            d=d+1
        if d <n: return False
        else: return True

def tong(a, b):
    sum=0
    for i in range(a, b+1):
        if isPrime(i):
            sum=sum+i
    return sum

while True:
    a=int(input("nhap so nguyen duong dau tien cua day:"))
    b=int(input("nhap so nguyen duong cuoi cua day:"))
    if a>b:
        print("Ban phai nhap a < b")
    else:
        break
sum = tong(a, b)
print(f"Tong cac so nguyen to trong doan [{a}, {b}] = {sum}")

```
